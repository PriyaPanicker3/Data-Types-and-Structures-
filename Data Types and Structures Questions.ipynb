#Data Types and Structures Questions

1. What are data structures, and why are they important ?

- Data structures are specialized ways of organizing, storing, and managing data in a computer so that it can be used efficiently.

 Data structures are important because they make data storage and access faster and more efficient, helping programs run better and solve problems effectively.

2. Explain the difference between mutable and immutable data types with examples

 Mutable data types can be changed after creation, while immutable data types cannot be changed once created.

 - Example:

Mutable: my_list = [1, 2]; my_list.append(3) → changes list

Immutable: my_str = "hi"; my_str[0] = "H" → error

3. What are the main differences between lists and tuples in Python

 In Python, lists and tuples store multiple items. Lists use [] and can be changed (mutable), while tuples use () and cannot be changed (immutable). Tuples are faster and used for fixed data, and lists are used when data needs to change.

4. Describe how dictionaries store data?

 Dictionaries in Python store data as key-value pairs. Each value is accessed using its unique key, not by position. They use curly braces {}

5. Why might you use a set instead of a list in Python ?

  Dictionaries in Python store data as key-value pairs using {}. Each key is unique and helps access its value quickly. They're useful for fast lookups and organized data storage.

6. What is a string in Python, and how is it different from a list ?
  
  In Python, a string is a sequence of characters enclosed in quotes, like "hello", while a list is a collection of items enclosed in square brackets, like [1, 2, "a"]. Strings are immutable, meaning they can’t be changed, but lists are mutable and can be modified

7. How do tuples ensure data integrity in Python ?

 Tuples in Python are immutable, which means their values cannot be changed once created. This protects the data from accidental changes, helping keep it safe and reliable. That's why tuples are good for storing constant or unchanging information.

8. What is a hash table, and how does it relate to dictionaries in Python ?

 A hash table stores data using key-value pairs and uses hashing to find data quickly. In Python, dictionaries use hash tables behind the scenes, which is why they are fast and great for storing and accessing data using keys.

9. Can lists contain different data types in Python ?

 Yes, lists can contain different data types in Python. A single list can hold integers, strings, floats, or even other lists.

10. Explain why strings are immutable in Python?

 In Python, strings are immutable, meaning you can't change their characters once created. This helps make programs safer and faster, as the string stays the same and doesn’t accidentally change during use

11. What advantages do dictionaries offer over lists for certain tasks ?

 Dictionaries are better than lists when you need fast access to data using names or labels (keys). They help keep data organized and easy to find without knowing the position, making them ideal for tasks like storing user info or settings.

12. Describe a scenario where using a tuple would be preferable over a list.

 A tuple is better than a list when you want to store data that should not change. For example, if you're storing the coordinates of a location (latitude, longitude), a tuple is best because the values stay fixed. Using a tuple helps protect the data from accidental changes.

13. How do sets handle duplicate values in Python ?

 Sets in Python do not allow duplicate values. If you add the same item more than once, it keeps only one copy. This helps keep your data unique automatically.

14. How does the “in” keyword work differently for lists and dictionaries ?

 In a list, the in keyword checks if a value exists. In a dictionary, it checks if a key exists, not the value. This helps quickly find items in both data types in different ways.

15. Can you modify the elements of a tuple? Explain why or why not.

 No, you cannot modify the elements of a tuple in Python because tuples are immutable. This means once a tuple is created, its values cannot be changed, added, or removed.


16. What is a nested dictionary, and give an example of its use case ?

 A nested dictionary is a dictionary inside another dictionary. It’s useful for storing detailed info, like student records with name, age, and grade. This helps keep related data grouped and easy to access.

17. Describe the time complexity of accessing elements in a dictionary.

 In Python, getting a value from a dictionary is very fast. It usually takes the same amount of time no matter how much data is inside, which is called O(1) time. This is why dictionaries are great for quick lookups.

18. In what situations are lists preferred over dictionaries?

 Lists are preferred over dictionaries when you need to store items in a specific order and access them by position (index). They are useful when the order of items matters or when you're working with simple sequences like numbers, names, or tasks. Lists are also easier to use when keys are not needed.

19. Why are dictionaries considered unordered, and how does that affect data retrieva ?

 Dictionaries were unordered in older Python versions, so item order didn’t matter. Since Python 3.7, they keep the insertion order, but data is still accessed using keys, not positions.
